```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [x for x in range(100)]
Y = [1 if x > 50 else 0 for x in (X)]
```

```python
plt.scatter(X,Y)
plt.show()
```

```python
learning_rate = 0.001
epochs = 600
w = 0
b = 0
n = len(X)
def sigmoid(x):
    return 1 / (1 + (np.exp(-x)))
for _ in range(epochs):
    w_deriv = 0
    b_deriv = 0
    for x,y in zip(X,Y):
        sigmoid_wxb = sigmoid(w*x+b)
        w_deriv += (-2 * (y - sigmoid_wxb) * (sigmoid_wxb * (1 - sigmoid_wxb)) * x) / n
        b_deriv += -2 * (y - sigmoid_wxb) * (sigmoid_wxb * (1 - sigmoid_wxb))
    w -= learning_rate * w_deriv
    b -= learning_rate * b_deriv
print('W:', w,'B:',b)
plt.scatter(X, Y, color='black')
range_X = range(min(X) - 1, max(X) + 1)
plt.plot(X, [sigmoid(w*x+b) for x in X], color='red')
plt.show()
```

```python
print('W:', w, 'B:', b)
predictions = [1 if sigmoid(w*x+b) >= 0.6 else 0 for x in X]
print(predictions, Y)
```

```python
accuracy = np.dot(Y, predictions)/(np.linalg.norm(Y)*np.linalg.norm(predictions))
print('Accuracy:',accuracy)
```
